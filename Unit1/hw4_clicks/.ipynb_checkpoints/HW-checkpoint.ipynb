# Практическое задание. Оценка качества поиска. Сплиты, асессоры.


**Дано**: файлы с1.txt и c2.txt, содержащие клики пользователей со сплита 1 и 2 соответственно.

**Формат**: каждая строка соответствует одной сессии и содержит десять цифр: нулей или единиц, обозначающих, был ли клик в документ в порядке возрастания ранга документа в поисковой выдаче.

**Задача**: Методом Bootstraping проверить гипотезу H0 о том, что CTR на двух сплитах можно объяснить шумами. Выполнить данный подсчет для любой другой кликовой метрики, например количество отказов (доля сессий без кликов), доля кликов в первый результат, позиция последнего клика, среднее количество кликов на сессию и так далее.

```python
import numpy as np
from tqdm import tqdm
from scipy import stats
```

```python
c1 = np.loadtxt("c1.txt")
c2 = np.loadtxt("c2.txt")
```

```python
c1.shape
```

Метрику CTR считают для сессий за определенный промежуток времени. Так как в исходных данных нет информации о времени, разделим сплиты на группы.

```python
def create_subsamples(c, num_split):
    
    '''
    c - np array
    num_split - array c will be divided into num_split equal arrays
    
    Returns list of arrays
    '''
    
    #creates replication of c array
    sub_sample = c[np.random.choice(c.shape[0], c.shape[0]), :]
    groups = np.split(sub_sample, num_split)
    
    return groups
```

```python
def get_ctrs(groups):
    
    '''
    Returns CTR values calculated for boostrap replicate of c array
    
    groups - list of arrays to calculate CTR
    '''
    # number of sessions
    session_count = groups[0].shape[0]
    CTRs = []
    for i in range(len(groups)):
        # number of session with no clicks
        zero_session_ct = np.count_nonzero(np.all((groups[i] == 0), axis=1))
        CTR = (session_count - zero_session_ct) / session_count
        CTRs.append(CTR)
    return CTRs
```

В качестве второй метрии выберем среднее количество кликов на сессию (Clicks per Query).

```python
def get_cpqs(groups):
    
    '''
    Returns CPQ (click per query) values calculated for boostrap replicate of c array
    '''

    # number of sessions
    session_count = groups[0].shape[0]
    CPQs = []
    for i in range(len(groups)):
        cpq = groups[i].sum(axis=1).mean()
        CPQs.append(cpq)

    return CPQs
```

Теперь сгенерируем реплики из наших данных, посчитаем t-статистику и определим, в скольких случах гипотеза H0 о равенстве параметров могла быть отвергнута на уровне значимости 5%.

```python
def cal_t_stat(ar1, ar2, n_iter, n_split):
    '''
    ar1 - np array
    ar2 - np array
    '''
    
    count_H1 = {}
    count_H1['CTR'] = 0
    count_H1['CPQ'] = 0
    
    deltas = {}
    deltas['CTR'] = []
    deltas['CPQ'] = []
    
    for i in tqdm(range(n_iter)):
        groups1 = create_subsamples(ar1, num_split)
        groups2 = create_subsamples(ar2, num_split)
        
        ctrs1 = get_ctrs(groups1)
        ctrs2 = get_ctrs(groups2)

        deltas['CTR'].append(np.mean(ctrs1) - np.mean(ctrs2))

        t, p_val = stats.ttest_ind(ctrs1,ctrs2,equal_var=False)
        if t > 1.96:
            count_H1['CTR'] += 1
            
        cpqs1 = get_cpqs(groups1)
        cpqs2 = get_cpqs(groups2)

        deltas['CPQ'].append(np.mean(cpqs1) - np.mean(cpqs2))

        t, p_val = stats.ttest_ind(cpqs1,cpqs2,equal_var=False)
        if t > 1.96:
            count_H1['CPQ'] += 1
    return count_H1, deltas
```

```python
count_H1, deltas = cal_t_stat(c1, c2, 500, 1000)
```

```python
print('CTR')
print('count H1 = {}, ASL = {}'.format(count_H1['CTR'], count_H1['CTR']/500))

print('CPQ')
print('count H1 = {}, ASL = {}'.format(count_H1['CPQ'], count_H1['CPQ']/500))
```

```python
from matplotlib import pyplot as plt

fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Difference between 1 and 2 groups')
ax1.plot(deltas['CPQ'])
ax1.set_title('CPQ')
ax2.plot(deltas['CTR'])
ax2.set_title('CTR')
```
